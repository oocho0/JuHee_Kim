```python
marriage_rate_g = pd.DataFrame(marriage_rate['2020'])
marriage_rate_g['total_2020']=total_pop['2020']
marriage_rate_g['인구감소지수 분류']=marriage_rate['인구감소지수 분류']
marriage_rate_g
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='darkgrid')

sns.scatterplot(x='total_2020', 
                y='2020', 
                hue='인구감소지수 분류', # different colors by group
                style='인구감소지수 분류', # different shapes by group
                s=80, # marker size
                data=marriage_rate_g)

plt.title('2020년 혼인률')
plt.xlabel('인구수')
plt.ylabel('혼인률')

plt.savefig('output/marriage_rate_2020.png')
plt.show()
```

```python
marriage_89 = marriage_rate[marriage_rate['인구감소지수 분류'].isin(['인구감소지역'])]
marriage_89
```

```python
marriage_18 = marriage_rate[marriage_rate['인구감소지수 분류'].isin(['관심지역'])]
marriage_18
```

```python
marriage_non = marriage_rate[marriage_rate['인구감소지수 분류'].isin(['안심지역'])]
marriage_non
```

```python
marriage_mean = pd.DataFrame({'인구감소지역':marriage_89.mean()})
marriage_mean['관심지역']=marriage_18.mean()
marriage_mean['안심지역']=marriage_non.mean()
marriage_mean['전국']=marriage_rate.mean()
marriage_mean
```

```python
df=marriage_mean.plot(kind='bar')

plt.title('5년간 결혼률의 변화')

plt.xticks(rotation=0)
plt.xlabel('연도')
plt.ylabel('결혼률의 평균')
lgd=df.legend(bbox_to_anchor=(1,1))

plt.savefig('output/marriage_5years_mean_bar.png',bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```

```python
plt.style.use('ggplot')
fig=plt.figure(figsize=(8,5))
ax=fig.add_subplot(1,1,1)

ax.plot(marriage_mean['인구감소지역'],marker='o',markerfacecolor='#ff5050',markersize=8,color='#ff5050',label='인구감소지역')
ax.plot(marriage_mean['관심지역'],marker='o',markerfacecolor='orange',markersize=8,color='orange',label='관심지역')
ax.plot(marriage_mean['안심지역'],marker='o',markerfacecolor='#00ccff',markersize=8,color='#00ccff',label='안심지역')
#ax.plot(marriage_mean['전국'],marker='o',markerfacecolor='indigo',markersize=8,color='skyblue',label='전국')

ax.legend(loc='best')

ax.set_title('5년간 혼인율의 변화')

ax.set_ylabel('혼인율의 평균')
ax.set_xlabel('연도')


plt.savefig('output/marriage_5years_mean_line.png')
plt.show()
```
