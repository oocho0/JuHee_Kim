```python
import pandas as pd

# 데이터프레임 변환
df = pd.read_excel('data/남북한발전전력량.xlsx', engine='openpyxl') 
```

```python
# 남한, 북한 발전량 합계 데이터만 추출
df_ns = df.iloc[[0, 5], 3:]

# 행 인덱스 변경            
df_ns.index = ['South','North']

# 열 이름의 자료형을 정수형으로 변경
df_ns.columns = df_ns.columns.map(int) 
```

```python
# 선 그래프 그리기
df_ns.plot()
```

```python
# 행, 열 전치
tdf_ns = df_ns.T
print(tdf_ns.head(), '\n','*'*40)

tdf_ns.plot()
```

```python
# 행, 열 전치
tdf_ns = df_ns.T
print(tdf_ns.head())

tdf_ns.plot(kind='bar')
```

```python
import pandas as pd
import seaborn as sns

df=sns.load_dataset('mpg')
```

```python
df.plot(kind='hist')
```

```python
# 2개의 열을 선택하여 산점도 그리기
df.plot(x='weight',y='mpg', kind='scatter')
```

```python
# 열을 선택하여 박스 플롯 그리기
df[['mpg','cylinders']].plot(kind='box')
```
