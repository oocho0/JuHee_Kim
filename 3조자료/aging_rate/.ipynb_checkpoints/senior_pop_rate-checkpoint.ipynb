```python
for i in ingu_89 :
    senior_rate.loc[i,'ingu_89'] = '인구감소지역'

senior_rate
```

```python
senior_rate_g = pd.DataFrame(senior_rate['2020'])
senior_rate_g['total_2020']=total_pop['2020']
senior_rate_g['ingu_89']=senior_rate['ingu_89']
senior_rate_g
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='darkgrid')

sns.scatterplot(x='total_2020', 
                y='2020', 
                hue='ingu_89', # different colors by group
                style='ingu_89', # different shapes by group
                s=100, # marker size
                data=senior_rate_g)

plt.title('2020년 고령화 비율')
plt.xlabel('인구수')
plt.ylabel('고령화 비율')

plt.savefig('output/senior_rate_2020.png')
plt.show()
```

```python
senior_89 = senior_rate[senior_rate['ingu_89'].isin(['인구감소지역'])]
senior_89
```

```python
senior_non = senior_rate[senior_rate['ingu_89'].isin(['비인구감소지역'])]
senior_non
```

```python
senior_mean = pd.DataFrame({'ingu89':senior_89.mean()})
senior_mean['non']=senior_non.mean()
senior_mean['total']=senior_rate.mean()
senior_mean
```

```python
senior_mean.plot(kind='bar')

plt.title('5년간 노령화 비율의 변화')

plt.xticks(rotation=0)
plt.xlabel('연도')
plt.ylabel('노령화 비율의 평균')
plt.legend(bbox_to_anchor=(1,1))

plt.savefig('output/senior_5years_mean_bar.png')
plt.show()
```

```python
plt.style.use('ggplot')
fig=plt.figure(figsize=(8,5))
ax=fig.add_subplot(1,1,1)

ax.plot(senior_mean['ingu89'],marker='o',markerfacecolor='olive',markersize=10,color='green',label='인구감소지역')
ax.plot(senior_mean['non'],marker='o',markerfacecolor='red',markersize=10,color='magenta',label='비인구감소지역')
ax.plot(senior_mean['total'],marker='o',markerfacecolor='indigo',markersize=10,color='skyblue',label='전체 지역')

ax.legend(loc='best')

ax.set_title('5년간 노령화 비율의 변화')

ax.set_ylabel('노령화 비율의 평균')
ax.set_xlabel('연도')
plt.legend(loc=6)

plt.savefig('output/senior_5years_mean_line.png')
plt.show()
```
