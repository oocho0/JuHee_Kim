```python
data=pd.read_csv('project.csv')
data
```

```python
pj_data = data.drop(['SIG_CD','SIG_KOR_NM','시도'], axis=1)
pj_data.set_index('시군구', inplace=True)
pj_data
```

```python
pj_data.columns
```

```python
from sklearn import preprocessing

label_encoder =preprocessing.LabelEncoder()
onehot_encoder = preprocessing.OneHotEncoder()
```

```python
onehot_under14 =label_encoder.fit_transform(pj_data['유소년비율'])
onehot_birth =label_encoder.fit_transform(pj_data['조출생률'])
onehot_oph =label_encoder.fit_transform(pj_data['1인 가구비율'])
onehot_tunnel = label_encoder.fit_transform(pj_data['터널수'])
onehot_bakery = label_encoder.fit_transform(pj_data['만명당 매장 수'])
onehot_hs = label_encoder.fit_transform(pj_data['만명 당 의료기관 수'])
onehot_hs_3 = label_encoder.fit_transform(pj_data['만명당 3차병원수'])
onehot_hs_2 = label_encoder.fit_transform(pj_data['만명당 2차병원 수'])
onehot_hs_1 = label_encoder.fit_transform(pj_data['만명당 1차병원 수'])
onehot_hs_h = label_encoder.fit_transform(pj_data['만명당 요양병원 수'])
onehot_hs_k = label_encoder.fit_transform(pj_data['만먕당 한의원 수'])
onehot_ids_1 = label_encoder.fit_transform(pj_data['1차산업비율'])
onehot_ids_2 = label_encoder.fit_transform(pj_data['2차산업비율'])
onehot_ktx = label_encoder.fit_transform(pj_data['ktx거리'])
onehot_house = label_encoder.fit_transform(pj_data['빈집비율'])
onehot_mountain = label_encoder.fit_transform(pj_data['산지면적/도시면적'])
onehot_money = label_encoder.fit_transform(pj_data['재정자립도'])
onehot_pop = label_encoder.fit_transform(pj_data['연평균인구증감률'])
```

```python
pj_data['under14']=onehot_under14
pj_data['birth']=onehot_birth
pj_data['oph']=onehot_oph
pj_data['tunnel']=onehot_tunnel
pj_data['bakery']=onehot_bakery
pj_data['hs']=onehot_hs
pj_data['hs_3']=onehot_hs_3
pj_data['hs_2']=onehot_hs_2
pj_data['hs_1']=onehot_hs_1
pj_data['hs_h']=onehot_hs_h
pj_data['hs_k']=onehot_hs_k
pj_data['ids_1']=onehot_ids_1
pj_data['ids_2']=onehot_ids_2
pj_data['ktx']=onehot_ktx
pj_data['house']=onehot_house
pj_data['mountain']=onehot_mountain
pj_data['money']=onehot_money
pj_data['pop']=onehot_pop
pj_data.iloc[:,18:]
```

```python
from sklearn import cluster

X=pj_data.iloc[:,18:]

X=preprocessing.StandardScaler().fit(X).transform(X)

dbm=cluster.DBSCAN(eps=0.2, min_samples=5)

dbm.fit(X)

cluster_label = dbm.labels_
```

```python
pj_data['Cluster']=cluster_label
```
