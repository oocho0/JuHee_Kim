```python
for i in ingu_89 :
    death_rate.loc[i,'인구감소지수 분류'] = '인구감소지역'
    
for i in semi_18 :
    death_rate.loc[i,'인구감소지수 분류']='관심지역'

death_rate
```

```python
deathRate_89 = death_rate[death_rate['인구감소지수 분류'].isin(['인구감소지역'])]
deathRate_89
```

```python
deathRate_18 = death_rate[death_rate['인구감소지수 분류'].isin(['관심지역'])]
deathRate_18
```

```python
deathRate_non = death_rate[death_rate['인구감소지수 분류'].isin(['안심지역'])]
deathRate_non
```

```python
death_mean = pd.DataFrame({'인구감소지역':deathRate_89.mean()})
death_mean['관심지역']=deathRate_18.mean()
death_mean['안심지역']=deathRate_non.mean()
#death_mean['전국']=death_rate.mean()
death_mean
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='darkgrid')

plt.style.use('ggplot')
fig=plt.figure(figsize=(8,5))
ax=fig.add_subplot(1,1,1)

ax.plot(death_mean['인구감소지역'],marker='o',markerfacecolor='#ff5050',markersize=8,color='#ff5050',label='인구감소지역')
ax.plot(death_mean['관심지역'],marker='o',markerfacecolor='orange',markersize=8,color='orange',label='관심지역')
ax.plot(death_mean['안심지역'],marker='o',markerfacecolor='#00ccff',markersize=8,color='#00ccff',label='안심지역')
#ax.plot(deathmean['전국'],marker='o',markerfacecolor='indigo',markersize=10,color='skyblue',label='전국')

ax.legend(loc='best')

ax.set_title('5년간 사망율의 변화')

ax.set_ylabel('사망율의 평균')
ax.set_xlabel('연도')
plt.legend(loc=6)

plt.savefig('output/deathmean_5years_line.png', bbox_inches='tight')
plt.show()
```

```python
import numpy as np

count, bin_dividers = np.histogram(deathRate_89['2020'],bins=4)
bin_dividers
```

```python
bin_names=['696.8~973','973.75~1250','1250.7~1527','1527.65~1805']

deathRate_89['hp_bin']=pd.cut(x=deathRate_89['2020'],bins=bin_dividers,labels=bin_names,include_lowest=True)
deathRate_89
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='whitegrid')

sns.scatterplot(x='pop', y='2020', hue='hp_bin',style='hp_bin',s=60,data=deathRate_89)

plt.title('인구감소지역 조출산율')
plt.xlabel('인구수')
plt.ylabel('조출산율')
plt.legend(fontsize=6)
plt.savefig('output/deathRate_rate_89_bins.png', bbox_inches='tight')
plt.show()
```
