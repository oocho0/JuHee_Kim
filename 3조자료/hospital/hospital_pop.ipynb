```python
hs_sum_89 = hs_sum[hs_sum['인구감소지수 분류'].isin(['인구감소지역'])]
hs_sum_89['sido']=sido['sido']
hs_sum_89['전체']=hs_sum_89['sum']/hs_sum_89['pop']*10000
hs_sum_89['삼차']=hs_sum_89['상급종합병원']/hs_sum_89['pop']*10000
hs_sum_89['이차']=hs_sum_89['종합병원']/hs_sum_89['pop']*10000
hs_sum_89['일차']=hs_sum_89['병원']/hs_sum_89['pop']*10000
hs_sum_89['요양']=hs_sum_89['요양병원']/hs_sum_89['pop']*10000
hs_sum_89['한의']=hs_sum_89['한의원']/hs_sum_89['pop']*10000
hs_sum_89.reset_index(inplace=True)
hs_sum_89
```
