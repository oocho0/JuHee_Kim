```python
korea_2016=hh['총가구_2016'].sum()
korea_2017=hh['총가구_2017'].sum()
korea_2018=hh['총가구_2018'].sum()
korea_2019=hh['총가구_2019'].sum()
korea_2020=hh['총가구_2020'].sum()
```

```python
hh_total=pd.DataFrame(hh['인구감소지수 분류'])
hh_total['2016']=hh['총가구_2016']
hh_total['2017']=hh['총가구_2017']
hh_total['2018']=hh['총가구_2018']
hh_total['2019']=hh['총가구_2019']
hh_total['2020']=hh['총가구_2020']
hh_total
```

```python
hh_total_89=hh_total[hh_total['인구감소지수 분류'].isin(['인구감소지역'])]
hh_total_89
```

```python
hh_total_18=hh_total[hh_total['인구감소지수 분류'].isin(['관심지역'])]
hh_total_non=hh_total[hh_total['인구감소지수 분류'].isin(['안심지역'])]
hh_total_mean=pd.DataFrame({'인구감소지역':hh_total_89.mean(),
                            '관심지역':hh_total_18.mean(),
                            '안심지역':hh_total_non.mean(),
                            '전국':hh_total.mean()})
hh_total_mean
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

rc('font', family='AppleGothic')

plt.style.use('ggplot')
fig=plt.figure(figsize=(8,5))
ax=fig.add_subplot(1,1,1)

ax.plot(hh_total_mean['인구감소지역'],marker='o',markerfacecolor='olive',markersize=8,color='green',label='인구감소지역')
ax.plot(hh_total_mean['관심지역'],marker='o',markerfacecolor='orange',markersize=8,color='yellow',label='관심지역')
ax.plot(hh_total_mean['안심지역'],marker='o',markerfacecolor='red',markersize=8,color='magenta',label='안심지역')
ax.plot(hh_total_mean['전국'],marker='o',markerfacecolor='indigo',markersize=8,color='skyblue',label='전국')

lgd=ax.legend(bbox_to_anchor=(1,1))
ax.set_title('5년간 가구수의 변화')
ax.set_ylabel('가구수의 평균')
ax.set_xlabel('연도')

plt.savefig('output/hh_5years_mean_line.png',bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
