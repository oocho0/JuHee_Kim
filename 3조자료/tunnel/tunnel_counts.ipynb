```python
tunnel_fin=tunnel_num[tunnel_num.index.str.contains('부산광역시 동구')]
for i in ingu_88 :
    tunnel_fin_i = tunnel_num[tunnel_num.index.str.contains(i)]
    tunnel_fin = pd.concat([tunnel_fin,tunnel_fin_i])
    
tunnel_fin.index
tunnel_fin_1 = pd.DataFrame({'부산광역시 영도구':[0],
                            '인천광역시 강화군':[0],
                            '인천광역시 옹진군':[0],
                            '충청남도 태안군':[0],
                            '전라남도 신안군':[0],
                            '전라남도 완도군':[0],
                            '전라남도 진도군':[0],
                            '전라남도 해남군':[0],
                            '경상북도 울진군':[0],
                            '경상남도 의령군':[0],
                            '경상남도 창녕군':[0]})
tunnel_fin_1=tunnel_fin_1.T
tunnel_fin_1.rename(columns={0:'sigungu'},inplace=True)
tunnel_fin_1
```

```python
tunnel_fin = pd.concat([tunnel_fin,tunnel_fin_1])
tunnel_fin.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False}, style='darkgrid')

sns.scatterplot(x='green', y='tunnel',s=40,data=tunnel_fin)

tunnel_fin.reset_index(inplace=True)
for i in range(len(tunnel_fin)):
    row=tunnel_fin.iloc[i]
    name=row['sigungu']
    x = row['green']
    y = row['tunnel']
    if name in ['강원도 홍천군','충청북도 옥천군']:
        plt.text(x-60000, y-5, name, fontsize=8)
tunnel_fin.set_index('sigungu', inplace=True)

plt.title('2020년 터널수-산지면적')
plt.xlabel('산지면적')
plt.ylabel('터널수')

plt.savefig('output/tunnel_2020.png', bbox_inches='tight')
plt.show()
```

```python
tunnel_fin['tunnel_rate']=tunnel_fin['tunnel']/tunnel_fin['green']*10000
tunnel_fin
```

```python
df1 = sns.scatterplot(x='green', y='tunnel_rate',s=60,data=tunnel_fin)

plt.title('2020년 터널수-산지면적')
plt.xlabel('산지면적')
plt.ylabel('터널수/산지면적 비율')
lgd=df1.legend(bbox_to_anchor=(1,1), fontsize=10)


plt.savefig('output/tunnel_rate_1_2020.png',bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
