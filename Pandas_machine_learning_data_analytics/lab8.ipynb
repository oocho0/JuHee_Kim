```python
import pandas as pd

#문제 1

df_1 = pd.read_csv('data/score.csv')
print("컬럼명 : ", df_1.columns, "\n인덱스 : ", df_1.index)

df_1_1=df_1.copy()
df_1_1.drop("name", axis=1, inplace=True)
total = df_1_1.sum(axis=1)
avg = df_1_1.mean(axis=1)
avg = avg.round(decimals=2)
df_1["total"] = total
df_1["avg"] = avg

display(df_1)
```

```python
import pandas as pd

#문제 2

df_2 = pd.read_csv("data/score_noheader.csv", header=None)
display(df_2)
print("컬럼명 : ", df_2.columns, "\n인덱스 : ", df_2.index)

print("컬럼명 추가 ", "-" * 10)
df_2.rename(columns={0:"name",
                     1: "kor",
                     2: "eng",
                     3: "mat"}, inplace=True)
print("컬럼명 : ", df_2.columns)
display(df_2)

print("total과 avg열 추가", "-"*10)
df_2_1=df_2.copy()
df_2_1.drop("name", axis=1, inplace=True)
total=df_2_1.sum(axis=1)
avg = df_1_1.mean(axis=1)
avg = avg.round(decimals=2)
df_2["total"] = total
df_2["avg"] = avg
display(df_2)
```
