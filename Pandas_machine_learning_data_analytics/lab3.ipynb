```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

categories = list()
titles = list()
urlstr = 'https://news.daum.net/'
r = requests.get(urlstr)
r.encoding = "utf-8"
bs = BeautifulSoup(r.text, 'html.parser')
a_cat = bs.select('span.txt_category')
a_tit = bs.select('strong.tit_g > a')
for i in range(0,20):
    categories.append((a_cat)[i].text)
    titles.append((a_tit)[i].text.strip())
daum_article = {'category' : categories, 'title' : titles}
df = pd.DataFrame(daum_article)
display(df)
df.to_csv('output/daum_article.csv')
```
