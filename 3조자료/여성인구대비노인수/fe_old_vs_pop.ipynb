```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc
%matplotlib inline

plt.rc('font', family='AppleGothic')
sns.set(font='AppleGothic', rc={'axes.unicode_minus':False},style='darkgrid')
data = pd.read_csv('data/fe_old_pop.csv')
data.count()
```

```python
data['인구성장률 2011-2021'].fillna(data['인구성장률 2011-2021'].mean(), inplace=True)
data.count()
```

```python
data.set_index(['시군구'], inplace=True)
data.head()
```

```python
fig=plt.figure(figsize=(8,6))


plt.title('기존 소멸위험지수와 인구성장률')

splot1=sns.regplot(data=data, x='소멸위험지수', y='인구성장률 2011-2021')
plt.xlabel('20~39세 여성인구/65세이상 인구')
plt.ylabel('2011-2021년 인구성장률')
plt.show()
sfig1 = splot1.get_figure()
sfig1.savefig('output/기존 소멸위험지수와 인구성장률2011.png', orientation='landscape')
```

```python
fig=plt.figure(figsize=(8,6))


plt.title('기존 소멸위험지수와 인구성장률')

splot2=sns.regplot(data=data, x='소멸위험지수', y='인구성장률 2016-2021')
plt.xlabel('20~39세 여성인구/65세이상 인구')
plt.ylabel('2016-2021년 인구성장률')
plt.show()
sfig2 = splot2.get_figure()
sfig2.savefig('output/기존 소멸위험지수와 인구성장률2016.png', orientation='landscape')
```
