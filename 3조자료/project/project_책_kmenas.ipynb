```python
data=pd.read_csv('project.csv')
data
```

```python
data.info()
```

```python
pj_data = data.drop(['SIG_CD','SIG_KOR_NM','시도'], axis=1)
pj_data.set_index('시군구', inplace=True)
pj_data
```

```python
df=pj_data.iloc[:,:]
```

```python
from sklearn import preprocessing
df = preprocessing.StandardScaler().fit(df).transform(df)
```

```python
from sklearn import cluster

kmeans=cluster.KMeans(init='k-means++',n_clusters=3, n_init=10)
```

```python
kmeans.fit(df)
```

```python
cluster_label=kmeans.labels_
cluster_label
```

```python
pj_data['cluster']=cluster_label
pj_data.columns
```

```python
pj_data.plot(kind='scatter', x='연평균인구증감률',y='만명당 매장 수', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='조출생률', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='1인 가구비율', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='만명 당 의료기관 수', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='유소년비율', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='터널수', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='1차산업비율', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='빈집비율', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='산지면적/도시면적', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
pj_data.plot(kind='scatter', x='연평균인구증감률',y='재정자립도', c='cluster', cmap='Set1', colorbar=False, figsize=(10,10))
```
