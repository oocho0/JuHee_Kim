```python
import urllib.request
import urllib.parse
import requests
from bs4 import BeautifulSoup
url = "http://unico2013.dothome.co.kr/crawling/exercise_bs.html"
req_html = requests.get(url)
html = req_html.content
html = html.decode("utf-8")
bs = BeautifulSoup(html, 'html.parser')
print((urllib.request.urlopen(url)).read(1094).decode("utf-8"))
print("="*120)
print("(1) <h1>의 content : ", bs.h1.text)
print("-"*120)
print("(2) text type인 <a>의 content : href 속성값 ", )
for i in range(3):
    a = bs.select("a")[i]
    b = bs.select("a[href]")[i]
    print(a.text," : ",b.attrs)
print("-"*120)
print("(3) <img>의 src 속성값 : ", bs.img['src'])
print("-"*120)
print("(4) 1st <h2> content인 <ul>의 자식 tags 중 style 속상값이 green으로 끝나는 tag의 content : ", bs.select('li[style$=green]')[0].text)
print("-"*120)
print("(5) 2nd <h2> content인 <ol>의 모든 자식 tags의 content : ", bs.ol.text)
print("-"*120)
print("(6) <table>의 모든 자손 tags의 content : ", bs.table.text)
print("-"*120)
print("(7) class 속성이 name인 <tr>의 content : ")
for i in range(len(bs.select('tr.name'))):
    c = bs.select('tr.name')[i]
    print(c.text)
print("-"*120)
print("(8) ID 속성이 target인 <td>의 content : ")
for i in range(len(bs.select('#target'))):
    d = bs.select('#target')[i]
    print(d.text)
```
