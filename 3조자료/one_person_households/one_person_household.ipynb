```python
oph_mean = pd.DataFrame({'인구감소지역':oph_89.mean()})
oph_mean['관심지역']=oph_18.mean()
oph_mean['안심지역']=oph_non.mean()
#oph_mean['전국']=oph_rate.mean()
oph_mean=oph_mean.T
oph_mean
```

```python
#oph_mean_g = pd.DataFrame({'전체':(oph_mean['전체_1인가구비율_2020']-oph_mean['전체_1인가구비율_2016'])})
oph_mean_g = pd.DataFrame({'미성년':(oph_mean['전체_1인가구_미성년_2020']-oph_mean['전체_1인가구_미성년_2016'])})
oph_mean_g['청년']=oph_mean['전체_1인가구_청년_2020']-oph_mean['전체_1인가구_청년_2016']
oph_mean_g['장년']=oph_mean['전체_1인가구_장년_2020']-oph_mean['전체_1인가구_장년_2016']
oph_mean_g['노년']=oph_mean['전체_1인가구_노년_2020']-oph_mean['전체_1인가구_노년_2016']

oph_mean_g = oph_mean_g.T
oph_mean_g
```

```python
plt.figure(figsize=(20,10))

df=oph_mean_g.plot(kind='bar')

plt.title('2016년과 2020년 1인가구비율 변화')
plt.xticks(rotation=0)
plt.xlabel('세대별')
plt.ylabel('2020년-2016년 1인가구비율')
lgd=df.legend(bbox_to_anchor=(1,1))

plt.savefig('output/one_person_household_mean_bar.png',bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
